```python
from form_funcs import *

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# Formatando a Tabela

```python
df = pd.read_csv('meetup1.csv', header = None)
```

```python
dict_cols = {2 :  'age',
             3 :  'dist',
             4 :  'area',
             5 :  'interests',
             6 :  'meet_interest',
             7 :  'meet_weekday',
             8 :  'meet_subject',
             9 :  'meet_type',
             10 : 'l_interest',
             11 : 'l_type',
             12 : 'l_weekday'}
```

```python
df = df.rename(columns = dict_cols)
```

Alguém conseguiu preencher o questionário só com `NaN`s.

```python
df = df.dropna(subset = ['age'])
```

Title Font Size

```python
title_font_size = 15
```

Figure Sizes

```python
size_figure = (7, 5)
```

```python
size_pie = (7, 7)
```

Images Path

```python
path_img = 'images/'
```

DPI

```python
dpi = 200
```

# Informações Gerais

## Idade

```python
age = df['age']
```

```python
plt.figure(figsize = size_figure)
plt.hist(age, width = 2.5)
plt.xlabel('idade')
plt.ylabel('# de pessoas')
plt.title('Histograma de Idades', fontsize = title_font_size)
plt.savefig(path_img + 'age.jpg', dpi = dpi)
plt.show()
```

## Área

```python
area = df['area']
```

```python
area_sizes = area.value_counts() / area.count()
```

```python
explode = tuple([0.05 for _ in range(0, len(area_sizes))])
```

```python
plt.figure(figsize = size_pie)
plt.pie(area_sizes, labels = area_sizes.index, autopct='%1.1f%%')
plt.title('Áreas de Atuação', fontsize = title_font_size)
plt.savefig(path_img + 'areas_de_atuacao.jpg', dpi = dpi)
plt.show()
```

## Interesses

```python
s, l = split_unique_plus_stats(df, 'interests')
```

```python
plot_sups(size_figure = size_figure,
          s = s, l = l,
          xlabel = '% de adesão',
          ylabel = 'Atividades',
          title = 'Interesses em relação a Meet Ups e Confraternizações\n (É possível assinalar mais de uma opção)',
          title_font_size = title_font_size,
          path = path_img,
          file_name = 'interesses_gerais.jpg')
```

### Interesses por Departamento

```python
int_dep = sups_loop(df, column = 'interests', column_restriction = 'area')
```

Alguns departamentos não têm amostras o suficiente para serem úteis.

```python
for k, v in int_dep.items():
    
    s = v[0]
    l = v[1]
    
    plot_sups(size_figure = size_figure,
              s = s, l = l,
              xlabel = '% de adesão',
              ylabel = 'Atividades',
              title = f'Interesses em relação a Meet Ups e Confraternizações\n do Departamento: {k}\n (É possível assinalar mais de uma opção)',
              title_font_size = title_font_size,
              path = path_img,
              file_name = f'interesses_gerais_{k}.jpg')
```
